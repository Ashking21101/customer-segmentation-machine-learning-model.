```python
# @title Handling Outliers with IQR


def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# remove outliers from Income and Total_Spend
df_clean = remove_outliers_iqr(df, 'Income')
df_clean = remove_outliers_iqr(df_clean, 'Total_Spend')
```

```python
# @title Feature Engineering & Preprocessing for Customer Segmentation


# Create New Features

# Total Children
df_clean['Total_Children'] = df_clean['Kidhome'] + df_clean['Teenhome']

# Customer tenure (how long they've been with the company)
from datetime import datetime
df_clean['Dt_Customer'] = pd.to_datetime(df_clean['Dt_Customer'])
df_clean['Customer_Since_Days'] = (datetime.now() - df_clean['Dt_Customer']).dt.days

# Average yearly spending
df_clean['Avg_Annual_Spend'] = df_clean['Total_Spend'] / (df_clean['Customer_Since_Days'] / 365 + 1)
```

```python
# @title  Select Features for Clustering


features_for_clustering = df_clean[[
    'Age', 'Income', 'Total_Children', 'Total_Spend',
    'Avg_Annual_Spend', 'Recency', 'NumWebPurchases',
    'NumCatalogPurchases', 'NumStorePurchases'
]]
```

We only want numeric features that make sense for segmentation (like age, income, spending). Avoid IDs or campaign responses.

```python
# @title Normalize the Data (StandardScaler)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_clustering)

# convert back to DataFrame
import pandas as pd
scaled_df = pd.DataFrame(scaled_features, columns=features_for_clustering.columns,index=features_for_clustering.index)
```

```python
# Model Building: KMeans Clustering for Customer Segmentation

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# 1. Determine the optimal number of clusters using the Elbow method
sse = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(K, sse, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# 2. Fit KMeans with the chosen number of clusters (for example, k=4)
optimal_k = 4  # Choose based on the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_df)
```

```python
joblib.dump(kmeans, 'model.pkl')
```

```python
# 3. Assign cluster labels back to the original DataFrame
df_clean['Cluster'] = cluster_labels
```

```python
# 4. Analyze clusters
cluster_summary = df_clean.groupby('Cluster')[features_for_clustering.columns].mean()
print(cluster_summary)
```
